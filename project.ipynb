#### Exercise 16. Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (you must use the subset obtained in Exercise 14) (★★★)

Print both values for each variable.

```python
# TODO
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# This CSV file contains semicolons instead of comas as separator
dataset = pd.read_csv('assets/real_estate.csv', sep=';')
hasSouthBelt = dataset["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])

filtered = dataset.loc[hasSouthBelt == True]

filteredMeanPrice = filtered.groupby('level5')["price"].mean()
filteredVarPrice = filtered.groupby('level5')["price"].var()

print("Mean of Price:\n")
print(filteredMeanPrice)
print("\n")
print("Variance of Price:\n")
print(filteredVarPrice)
print("\n")

filteredMeanRooms = filtered.groupby('level5')["rooms"].mean()
filteredVarRooms = filtered.groupby('level5')["rooms"].var()

print("Mean of Rooms:\n")
print(filteredMeanRooms)
print("\n")
print("Variance of Rooms:\n")
print(filteredVarRooms)
print("\n")

filteredMeanSurface = filtered.groupby('level5')["surface"].mean()
filteredVarSurface = filtered.groupby('level5')["surface"].var()

print("Mean of Surface Area:\n")
print(filteredMeanSurface)
print("\n")
print("Variance of Surface Area:\n")
print(filteredVarSurface)
print("\n")

filteredMeanBath = filtered.groupby('level5')["bathrooms"].mean()
filteredVarBath = filtered.groupby('level5')["bathrooms"].var()

print("Mean of Rooms:\n")
print(filteredMeanBath)
print("\n")
print("Variance of Rooms:\n")
print(filteredVarBath)
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# TODO
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# This CSV file contains semicolons instead of comas as separator
dataset = pd.read_csv('assets/real_estate.csv', sep=';')
hasSouthBelt = dataset["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])

filtered = dataset.loc[hasSouthBelt == True]

normalPrices = filtered.groupby("level5")["price"].apply(lambda x: (x - x.mean()) / x.std())

fig, axis = plt.subplots(4, 1)

axis[0].hist(normalPrices["Fuenlabrada"])
axis[1].hist(normalPrices["Leganés"])
axis[2].hist(normalPrices["Getafe"])
axis[3].hist(normalPrices["Alcorcón"])
plt.tight_layout()
plt.show()
```

The normalized distribution of housing prices is similar across all locations

```python
# TODO
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# This CSV file contains semicolons instead of comas as separator
dataset = pd.read_csv('assets/real_estate.csv', sep=';')
hasSouthBelt = dataset["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])

filtered = dataset.loc[hasSouthBelt == True]

prices = filtered.groupby("level5")["price"].mean()
bathrooms = filtered.groupby("level5")["bathrooms"].mean()

fig, axis = plt.subplots(4)

axis[0].scatter(bathrooms["Fuenlabrada"], prices["Fuenlabrada"])
axis[1].scatter(bathrooms["Leganés"], prices["Fuenlabrada"])
axis[2].scatter(bathrooms["Getafe"], prices["Fuenlabrada"])
axis[3].scatter(bathrooms["Alcorcón"], prices["Fuenlabrada"])
plt.tight_layout()
plt.show()
```

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ipyleaflet import Map, Marker, basemaps

# This CSV file contains semicolons instead of comas as separator
dataset = pd.read_csv('assets/real_estate.csv', sep=';')
hasSouthBelt = dataset["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])

filtered = dataset.loc[hasSouthBelt == True]

map = Map(center=(40.3285, -3.7535), zoom=11, basemap=basemaps.OpenStreetMap.Mapnik)
groupLat = filtered.groupby("level5")['latitude'].idxmax()

for level, index in groupLat.items():
    group = filtered.loc[index]
    marker = Marker(
        location=(float(group['latitude'].replace(',', '.')[:15]), float(group['longitude'].replace(',', '.')[:15])),
        draggable=False,
        title=group['level5'], 
        color= "green", 
        visible = True
    )
    map.add_layer(marker)

    

# Display the map
map
```
